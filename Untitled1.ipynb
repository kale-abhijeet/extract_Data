```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # For RSI calculation

# --- Configuration ---
RSI_PERIOD = 14
RSI_OVERBOUGHT = 70
RSI_OVERSOLD = 30
# For our "buy" suggestion, we'll look for RSI not being overbought, and ideally trending up or oversold.
RSI_BUY_THRESHOLD = 50 # If RSI < 50, it's not overbought. We can refine this.

# --- Helper for RSI ---
def calculate_rsi(data, period=RSI_PERIOD):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def analyze_stock(ticker_symbol, plot_charts=True):
    print(f"\n--- Analyzing {ticker_symbol} ---")
    try:
        # Append '.NS' for NSE-listed stocks if not already present
        if not ticker_symbol.endswith(".NS"):
            ticker_name = f"{ticker_symbol}.NS"
        else:
            ticker_name = ticker_symbol
            
        ticker = yf.Ticker(ticker_name)

        # Fetch historical data for the past 6 months
        hist = ticker.history(period="1y") # Using 1 year for better MA and RSI calculation

        if hist.empty:
            print(f"No historical data found for {ticker_symbol}. Skipping.")
            return None

        # Calculate 20-day and 50-day moving averages
        hist['MA20'] = hist['Close'].rolling(window=20).mean()
        hist['MA50'] = hist['Close'].rolling(window=50).mean()
        hist['RSI'] = calculate_rsi(hist, RSI_PERIOD)
        hist['Volume_MA20'] = hist['Volume'].rolling(window=20).mean()


        # --- Basic Signal Generation ---
        # (This is highly simplified and for illustrative purposes)
        signal_score = 0
        reasons = []

        # Check latest data points
        latest_close = hist['Close'].iloc[-1] if not hist['Close'].empty else np.nan
        latest_ma20 = hist['MA20'].iloc[-1] if not hist['MA20'].empty else np.nan
        latest_ma50 = hist['MA50'].iloc[-1] if not hist['MA50'].empty else np.nan
        latest_rsi = hist['RSI'].iloc[-1] if not hist['RSI'].empty else np.nan
        latest_volume = hist['Volume'].iloc[-1] if not hist['Volume'].empty else np.nan
        avg_volume = hist['Volume_MA20'].iloc[-1] if not hist['Volume_MA20'].empty else np.nan
        
        # 1. Golden Cross (MA20 > MA50)
        if pd.notna(latest_ma20) and pd.notna(latest_ma50) and latest_ma20 > latest_ma50:
            signal_score += 2
            reasons.append("MA20 above MA50 (Golden Cross tendency)")
            # Check if it just crossed recently
            if len(hist) > 51 and hist['MA20'].iloc[-2] <= hist['MA50'].iloc[-2]:
                 reasons[-1] += " - Recent Crossover!"
                 signal_score +=1 # Extra point for recent crossover

        # 2. RSI
        if pd.notna(latest_rsi):
            if latest_rsi < RSI_OVERSOLD:
                signal_score += 2
                reasons.append(f"RSI ({latest_rsi:.2f}) is Oversold (<{RSI_OVERSOLD})")
            elif latest_rsi < RSI_BUY_THRESHOLD:
                signal_score += 1
                reasons.append(f"RSI ({latest_rsi:.2f}) is below {RSI_BUY_THRESHOLD} (Not Overbought)")

        # 3. Volume Confirmation (Simple: current volume > 20-day avg volume)
        if pd.notna(latest_volume) and pd.notna(avg_volume) and latest_volume > avg_volume * 1.2: # 20% above average
            signal_score += 0.5
            reasons.append("Recent volume above 20-day average")

        # 4. Price above MA20 (short-term uptrend)
        if pd.notna(latest_close) and pd.notna(latest_ma20) and latest_close > latest_ma20:
            signal_score += 1
            reasons.append("Price above MA20 (short-term bullish)")

        info = ticker.info
        pe_ratio = info.get('trailingPE', np.nan)
        current_price = info.get('currentPrice', info.get('previousClose', latest_close)) # More robust way to get price

        # 5. P/E Ratio (very subjective, avoid if too high for general consideration)
        if pd.notna(pe_ratio) and pe_ratio < 35 and pe_ratio > 0: # Example: not excessively high and positive
            signal_score += 0.5
            reasons.append(f"P/E Ratio ({pe_ratio:.2f}) is reasonable")
        elif pd.notna(pe_ratio) and pe_ratio <=0:
            reasons.append(f"P/E Ratio ({pe_ratio:.2f}) is negative or zero")


        if plot_charts:
            plt.figure(figsize=(14, 10))
            
            # Price and MAs
            ax1 = plt.subplot(3, 1, 1)
            plt.plot(hist.index, hist['Close'], label='Closing Price', color='blue')
            plt.plot(hist.index, hist['MA20'], label='20-Day MA', color='orange')
            plt.plot(hist.index, hist['MA50'], label='50-Day MA', color='green')
            plt.title(f"{info.get('shortName', ticker_symbol)} Stock Price and Moving Averages")
            plt.ylabel("Price (INR)")
            plt.legend()
            plt.grid(True)

            # Volume
            ax2 = plt.subplot(3, 1, 2, sharex=ax1)
            plt.bar(hist.index, hist['Volume'], label='Volume', color='grey', alpha=0.7)
            plt.plot(hist.index, hist['Volume_MA20'], label='Volume 20-Day MA', color='black')
            plt.ylabel("Volume")
            plt.legend()
            plt.grid(True)

            # RSI
            ax3 = plt.subplot(3, 1, 3, sharex=ax1)
            plt.plot(hist.index, hist['RSI'], label='RSI', color='purple')
            plt.axhline(RSI_OVERBOUGHT, color='red', linestyle='--', label=f'Overbought ({RSI_OVERBOUGHT})')
            plt.axhline(RSI_OVERSOLD, color='green', linestyle='--', label=f'Oversold ({RSI_OVERSOLD})')
            plt.axhline(RSI_BUY_THRESHOLD, color='blue', linestyle=':', label=f'Mid-Point ({RSI_BUY_THRESHOLD})')
            plt.title("Relative Strength Index (RSI)")
            plt.ylabel("RSI")
            plt.xlabel("Date")
            plt.legend()
            plt.grid(True)
            
            plt.tight_layout()
            plt.show()

        # Display latest financial metrics
        print(f"Company: {info.get('longName', 'N/A')}")
        print(f"Sector: {info.get('sector', 'N/A')}")
        print(f"Industry: {info.get('industry', 'N/A')}")
        print(f"Current Price: {current_price if pd.notna(current_price) else 'N/A'}")
        print(f"Market Cap: {info.get('marketCap', 'N/A')}")
        print(f"PE Ratio (TTM): {pe_ratio if pd.notna(pe_ratio) else 'N/A'}")
        print(f"Dividend Yield: {info.get('dividendYield', 'N/A')*100 if info.get('dividendYield') else 'N/A'}%")
        print(f"52 Week High: {info.get('fiftyTwoWeekHigh', 'N/A')}")
        print(f"52 Week Low: {info.get('fiftyTwoWeekLow', 'N/A')}")
        print(f"Latest MA20: {latest_ma20:.2f}" if pd.notna(latest_ma20) else "Latest MA20: N/A")
        print(f"Latest MA50: {latest_ma50:.2f}" if pd.notna(latest_ma50) else "Latest MA50: N/A")
        print(f"Latest RSI: {latest_rsi:.2f}" if pd.notna(latest_rsi) else "Latest RSI: N/A")
        
        print(f"\nSignal Score: {signal_score}")
        if reasons:
            print("Potential Buy Reasons:")
            for reason in reasons:
                print(f"- {reason}")
        else:
            print("No strong bullish signals detected based on current criteria.")
            
        return {
            "ticker": ticker_symbol,
            "name": info.get('shortName', ticker_symbol),
            "price": current_price,
            "score": signal_score,
            "reasons": reasons,
            "pe_ratio": pe_ratio,
            "sector": info.get('sector', 'N/A'),
            "ma20_above_ma50": latest_ma20 > latest_ma50 if pd.notna(latest_ma20) and pd.notna(latest_ma50) else False,
            "rsi_ok": latest_rsi < RSI_BUY_THRESHOLD if pd.notna(latest_rsi) else False
        }

    except Exception as e:
        print(f"Could not analyze {ticker_symbol}: {e}")
        return None

def suggest_stocks_from_pool(stock_pool, capital=10000, num_suggestions=5, min_score_threshold=2.0):
    print(f"\n--- Analyzing Stock Pool for Investment (Capital: INR {capital}) ---")
    print("IMPORTANT: This is NOT financial advice. Do your own research.\n")

    analyzed_stocks = []
    for ticker_symbol in stock_pool:
        analysis = analyze_stock(ticker_symbol, plot_charts=False) # Don't plot for each in pool initially
        if analysis and pd.notna(analysis['price']) and analysis['price'] > 0: # Ensure price is valid
            analyzed_stocks.append(analysis)
        print("-" * 30) # Separator

    if not analyzed_stocks:
        print("No stocks could be analyzed from the pool.")
        return

    # Filter stocks that meet a minimum score and are affordable for at least one share
    # (assuming we want to buy at least one share per suggested stock)
    
    # Sort by score (descending)
    potential_buys = sorted(
        [s for s in analyzed_stocks if s['score'] >= min_score_threshold], 
        key=lambda x: x['score'], 
        reverse=True
    )

    if not potential_buys:
        print(f"No stocks met the minimum signal score threshold of {min_score_threshold}.")
        return

    print(f"\n--- Top {num_suggestions} Potential Stock Suggestions (based on score >= {min_score_threshold}) ---")
    
    suggested_count = 0
    capital_per_stock = capital / num_suggestions if num_suggestions > 0 else capital

    for i, stock in enumerate(potential_buys):
        if suggested_count >= num_suggestions:
            break

        print(f"\nSuggestion #{suggested_count + 1}: {stock['name']} ({stock['ticker']})")
        print(f"  Current Price: INR {stock['price']:.2f}")
        print(f"  Signal Score: {stock['score']:.1f}")
        print(f"  P/E Ratio: {stock['pe_ratio']:.2f}" if pd.notna(stock['pe_ratio']) else "  P/E Ratio: N/A")
        print(f"  Sector: {stock['sector']}")
        print(f"  Reasons for potential interest:")
        for reason in stock['reasons']:
            print(f"    - {reason}")
        
        if stock['price'] > 0:
            shares_possible_within_budget = int(capital_per_stock / stock['price'])
            if shares_possible_within_budget > 0:
                print(f"  With INR {capital_per_stock:.2f} (budget per stock): You could notionally buy ~{shares_possible_within_budget} shares.")
                suggested_count += 1
            else:
                print(f"  This stock (at INR {stock['price']:.2f}) is too expensive for the allocated budget per stock (INR {capital_per_stock:.2f}).")
        else:
            print(f"  Could not determine valid price for {stock['ticker']}.")
        
        # Optionally, plot for the top suggestions
        if suggested_count <= num_suggestions: # To avoid re-plotting if already suggested
             analyze_stock(stock['ticker'], plot_charts=True)


    if suggested_count == 0:
        print("\nNo stocks from the filtered list were affordable within the per-stock budget for the top suggestions.")
    elif suggested_count < num_suggestions:
        print(f"\nFound {suggested_count} suggestions that fit the criteria and budget.")

# --- Example Usage ---

# Analyze a single stock with plots
# analyze_stock('RELIANCE')
# analyze_stock('INFY')
# analyze_stock('TATAMOTORS')


# Suggest from a pool
# For NSE stocks, you generally don't need ".NS" if yfinance can infer it,
# but being explicit (like f"{ticker_symbol}.NS") is safer.
# The function now handles adding .NS if it's missing.
my_stock_pool = [
    'RELIANCE', 'TCS', 'INFY', 'HDFCBANK', 'ICICIBANK', 
    'SBIN', 'BAJFINANCE', 'KOTAKBANK', 'LT', 'MARUTI',
    'ITC', 'ASIANPAINT', 'NESTLEIND', 'POWERGRID', 'ULTRACEMCO' # Example Nifty 50 stocks
]
# You can add more diverse stocks, including mid-cap or small-cap, for broader analysis
# my_stock_pool.extend(['IRCTC', 'DMART', 'JUBLFOOD', 'IEX'])


suggest_stocks_from_pool(my_stock_pool, capital=10000, num_suggestions=5, min_score_threshold=2.5)

# To test with a stock that might not have a good score:
# analyze_stock('YESBANK') # Often has complex factors not captured by simple MAs/RSI
```
